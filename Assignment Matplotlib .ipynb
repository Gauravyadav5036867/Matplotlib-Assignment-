```python
#Question 

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is widely used in data science, machine learning, and scientific computing for its ability to produce high-quality plots and graphs. Matplotlib's pyplot module is particularly popular because it provides a MATLAB-like interface for creating and customizing visualizations easily.

Why is Matplotlib Used?
Versatility: It can produce a wide range of plots and graphs, including line plots, scatter plots, bar charts, histograms, and more.
Customization: Offers extensive customization options for colors, labels, annotations, and other plot elements.
Integration: Easily integrates with other libraries like NumPy, Pandas, and SciPy, making it ideal for scientific computing and data analysis.
Output Options: Supports multiple output formats, including PNG, PDF, SVG, and more, allowing for high-quality publication-ready figures.
Community and Documentation: Has a large, active community and extensive documentation, providing numerous resources and examples for users.
```

```python
#Question 2

import matplotlib.pyplot as plt
import numpy as np


np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title("Scatter Plot of Generated Data")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

```python
#Question 3

import matplotlib.pyplot as plt
import numpy as np


x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title("Line Plot 1")
plt.xlabel("x-axis")
plt.ylabel("y-axis")


plt.subplot(2, 2, 2)
plt.plot(x2, y2, 'r')
plt.title("Line Plot 2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")


plt.subplot(2, 2, 3)
plt.plot(x3, y3, 'g')
plt.title("Line Plot 3")
plt.xlabel("x-axis")
plt.ylabel("y-axis")


plt.subplot(2, 2, 4)
plt.plot(x4, y4, 'b')
plt.title("Line Plot 4")
plt.xlabel("x-axis")
plt.ylabel("y-axis")


plt.tight_layout()


plt.show()
```

```python
#Question 4

A bar plot (or bar chart) is a graphical representation of data where individual bars represent different categories or values. The length or height of each bar is proportional to the value it represents. Bar plots are used to compare different categories or groups, making it easy to visualize differences in data, trends, and patterns.

Why is a Bar Plot Used?
Comparison: To compare different categories or groups side by side.
Clarity: To provide a clear visual distinction between different values.
Trends: To display trends over time or across categories.
Simplicity: To present data in a straightforward, easy-to-understand manner.


import matplotlib.pyplot as plt
import numpy as np


company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])


plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.bar(company, profit, color='blue')
plt.title("Company Profits - Bar Plot")
plt.xlabel("Company")
plt.ylabel("Profit (in millions)")


plt.subplot(1, 2, 2)
plt.barh(company, profit, color='green')
plt.title("Company Profits - Horizontal Bar Plot")
plt.xlabel("Profit (in millions)")
plt.ylabel("Company")


plt.tight_layout()


plt.show()
```

```python
#Question 5

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of data based on a five-number summary: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. It provides a summary of the data's central tendency and spread, highlighting outliers and skewness in the distribution.

Why is a Box Plot Used?
Visualizing Distribution: It shows the spread and skewness of data.
Identifying Outliers: It identifies outliers in the data.
Comparing Distributions: It compares multiple distributions side by side.
Summarizing Data: It provides a concise summary of data distribution.


import matplotlib.pyplot as plt
import numpy as np

# Generating data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Creating a box plot
plt.figure(figsize=(8, 6))

# Combine data into a list
data = [box1, box2]


plt.boxplot(data, labels=['Box 1', 'Box 2'])


plt.title('Box Plot of box1 and box2')
plt.xlabel('Box')
plt.ylabel('Value')


plt.show()
```
